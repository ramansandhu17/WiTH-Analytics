```python
import pandas as pd

linkedin_df=pd.read_csv('linkedinposts.csv',skiprows=1)

linkedin_df.head()
```

```python
linkedin_df.info()

linkedin_df['Created date']=pd.to_datetime(linkedin_df['Created date'])

linkedin_df.info()
```

```python
#check if there are any nulls

linkedin_df.isnull().sum()
```

```python
columns=['Campaign name','Campaign start date','Campaign end date','Views','Offsite Views','Follows','Content Type']

linkedin_df.drop(columns=columns,axis=1,inplace=True)

linkedin_df.head()
```

```python
linkedin_df.isnull().sum()
```

```python
linkedin_df.columns=(linkedin_df.columns.str.strip().str.lower().str.replace(' ','_'))

linkedin_df.info()
```

```python
linkedin_df.to_csv('linkedin_cleaned.csv',index=False)
```
